```python
import sklearn as sk
from collections import Counter
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import random
```

```python
# Set num of students
num_students = 4000
faculty = "Engineering"
student_data = []
faculties = {
    "Engineering & Information Technology":
}

credit_blocks = {
    (0, 24): ["Mathematics one", "Introduction to Engineering Projects", "Business Requirements Modelling", "Programming 1"],
    (25, 48): ["Mathematics two", "Programming two", "Systems Testing and Quality Management", "Database Fundamentals"],
    (49, 72): ["Professional Practice Preparation 1", "Physical Modelling", "Data Structures and Algorithms", "Information System Development Methodologies","Software Engineering Elective one"],
    (73, 96): ["Engineering Project Appraisal", "Software Design Studio", "Software Engineering Elective two"],
    (97, 120): ["Designing Sustainable Engineering Projects", " Software Development Studio", "Software Engineering Elective Three"],
    (121, 144): ["Engineering Work Experience", "Professional Engineering Communication", "Software Analysis Studio", "Software Architecture"],
    (145, 168): ["Engineering Research Preparation", "Collaboration in Complex Projects", "Professional Experience Review", "Software Engineering Elective Four"],
    (169, 192): ["Engineering Capstone", "Software Innovation Studio", "Software Engineering Elective Five"]
    }

# Generate student data
for student_id in range(1, num_students + 1):
    student_name = f"Student {student_id}"
    course = random.choice(["Software Engineering"])

    # Generate credit points in clumps of 6
    credit_points = random.randint(0, 15) * 6
    chosen_subjects = []
    for student_id in range(4):  # Repeat the process four times
        for credit_range, subjects in credit_blocks.items():
            if credit_range[0] <= credit_points <= credit_range[1]:
                chosen_subject = random.choice(subjects)
                chosen_subjects.append(chosen_subject)
                break

        student_data.append([student_id, student_name, course, credit_points, chosen_subjects])

# Create a DataFrame
columns = ["StudentID", "StudentName", "Course", "CreditPoints", "ChosenSubjects"]
df = pd.DataFrame(student_data, columns=columns)

# Print the first few rows of the dataset
print(df.tail(25))
```

```python
X = df[['CreditPoints']]

k = 5
knn = NearestNeighbors(n_neighbors=k)
knn.fit(X)

student_idx = 0  # Replace with the index of the student you want to recommend courses for
distances, indices = knn.kneighbors(X.iloc[[student_idx]])

# Get the nearest students' indices and find their chosen subjects
nearest_students = df.iloc[indices[0]]

# Combine the chosen subjects of the nearest students
combined_subjects = []
for _, student in nearest_students.iterrows():
    combined_subjects.extend(student['ChosenSubjects'])

# Find the most common courses among the nearest students
recommended_courses = [subject for subject, count in Counter(combined_subjects).most_common()]

print(f"Recommended courses for Student {student_idx + 1}:")
print(recommended_courses)
```
